```python
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
a=pd.read_csv("income.csv")
a
```

```python
plt.scatter(a.Age,a['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

##  K MEANS CLUSTERING ALGORITHM

```python
# K MEANS CLUSTERING ALGORITHM

# steps to follow
# first find how many clusters there from the plot
# here 3 clusters
#
```

```python
b=KMeans(n_clusters=3)                                         # here 3 clusters
c=b.fit_predict(a[['Age','Income($)']])
c
```

```python
a['cluster']=c
a
```

```python
b.cluster_centers_
```

```python
d0=a[a.cluster==0]
d1=a[a.cluster==1]
d2=a[a.cluster==2]
plt.scatter(d0.Age,d0['Income($)'],color='g')
plt.scatter(d1.Age,d1['Income($)'],color='brown')
plt.scatter(d2.Age,d2['Income($)'],color='orange')
plt.scatter(b.cluster_centers_[:,0],b.cluster_centers_[:,1],color='blue',marker='*',label='centroid')
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend(title='Marking')
```

```python
# note that the above plot need to be scaled
# centroids are not correctly plotted
# here Scaleing using MinMaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
sclr=MinMaxScaler()                                     # creating object
sclr.fit(a[['Income($)']])
a['Income($)']=sclr.transform(a[['Income($)']])
sclr.fit(a[['Age']])
a['Age']=sclr.transform(a[['Age']])
a
```

```python
# now repeats K mEans CLUSTERING ALGORITHM
```

```python
b=KMeans(n_clusters=3)
c=b.fit_predict(a[['Age','Income($)']])
c
```

```python
a['cluster']=c
a
```

```python
b.cluster_centers_
```

```python
d0=a[a.cluster==0]
d1=a[a.cluster==1]
d2=a[a.cluster==2]
plt.scatter(d0.Age,d0['Income($)'],color='violet')
plt.scatter(d1.Age,d1['Income($)'],color='brown')
plt.scatter(d2.Age,d2['Income($)'],color='orange')
plt.scatter(b.cluster_centers_[:,0],b.cluster_centers_[:,1],color='blue',marker='*',label='centroid')
plt.xlabel('x label')
plt.ylabel('y albel')
plt.legend(title='MarKings')
```

```python
sse=[]
r=range(1,10)
for k in r:
    km=KMeans(n_clusters=k)
    km.fit(a[['Age','Income($)']])
    sse.append(km.inertia_)

# sse =sum of squared error
```

```python
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(r,sse)

# elbow plot
# to find k(no.of clusters) using elbow method
# here the lower bend is at 3, so k=3
```
