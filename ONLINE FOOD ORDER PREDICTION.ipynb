```python
x=d.iloc[:,0:53]
x.head()
```

```python
y=d.iloc[:,53]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/4,random_state=1) 
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
# 1. LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression(random_state = 0)
lgr.fit(x_train, y_train)

lgr_y_predict = lgr.predict(x_test)
print("lgr_y_predict :", lgr_y_predict)

lgr_score = metrics.accuracy_score(y_test, lgr_y_predict)
print("lgr_score     :", lgr_score)
print("r2_score      :", r2_score(y_test, lgr_y_predict))
```

```python
# 2. DECISION TREE CLASSIFIER

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 0)
dtc.fit(x_train, y_train)

dtc_y_predict = dtc.predict(x_test)
print("dtc_y_predict :", dtc_y_predict)

dtc_score = metrics.accuracy_score(y_test, dtc_y_predict)
print("dtc_score     :", dtc_score)
print("r2_score      :", r2_score(y_test, dtc_y_predict))
```

```python
# 3. RANDOM FOREST CLASSIFIER

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 7, random_state = 0)
rfc.fit(x_train, y_train)

rfc_y_predict = rfc.predict(x_test)
print("rfc_y_predict :", rfc_y_predict)

rfc_score = metrics.accuracy_score(y_test, rfc_y_predict)
print("rfc_score     :", rfc_score)
print("r2_score      :", r2_score(y_test, rfc_y_predict))
```

```python
# 4. SVM CLASSIFIER

from sklearn.svm import SVC
svc = SVC(kernel = 'linear')
svc.fit(x_train, y_train)

svc_y_predict = svc.predict(x_test)
print("svc_y_predict :", svc_y_predict)

svc_score = metrics.accuracy_score(y_test, svc_y_predict)
print("svc_score     :", svc_score)
print("r2_score      :", r2_score(y_test, svc_y_predict))
```

```python
# 5. NAIVE BAYES CLASSIFIER

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)

nb_y_predict = nb.predict(x_test)
print("nb_y_predict :", nb_y_predict)

nb_score = metrics.accuracy_score(y_test, nb_y_predict)
print("nb_score     :", nb_score)
print("nb_score      :", r2_score(y_test, nb_y_predict))
```

```python
# 6. KNN CLASSIFIER

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)

knn_y_predict = knn.predict(x_test)
print("knn_y_predict :", knn_y_predict)

knn_score = metrics.accuracy_score(y_test, knn_y_predict)
print("knn_score     :", knn_score)
print("r2_score      :", r2_score(y_test, knn_y_predict))
```

```python
comparison = pd.DataFrame({
    "ML Models":["Logistic Regression","Decision Tree Classifier","Random Forest Classifier","SVM Classifier","Naive Bayes Classifier","KNN Classifier"],
    "Accuracy Score":[lgr_score,dtc_score,rfc_score,svc_score,nb_score,knn_score]
})

sns.barplot(x = "Accuracy Score", y = "ML Models", data = comparison)
comparison.sort_values(by = "Accuracy Score", ascending = False)
```
